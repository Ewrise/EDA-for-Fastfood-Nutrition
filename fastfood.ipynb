### **Sayısal sütunların görselleştirilmesi**

Sayısal sütunların açıklayıcı istatistiklerini gösterin.

Sayısal sütunların minimum, maksimum, ortalama, medyan ve mod değerlerini gösterin.

Sayısal sütunların dağılımlarını histogram ile görselleştirin.

Sayısal sütunların kutu grafiğini (boxplot) görselleştirin.

Sayısal sütunların yoğunluğunu gösterin.

```python
def show_distribution(col,title):
    

    min_val = col.min()
    max_val = col.max()
    mean_val = col.mean()
    med_val = col.median()
    mod_val = col.mode()[0]

    print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(min_val,
                                                                                            mean_val,
                                                                                            med_val,
                                                                                            mod_val,
                                                                                            max_val))

   
    fig, ax = plt.subplots(2, 1, figsize = (15,10))

    
    ax[0].hist(col)
    ax[0].set_ylabel('Values',fontsize=10)

    
    ax[0].axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2, label='Minimum')
    ax[0].axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2, label='Mean')
    ax[0].axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2, label='Median')
    ax[0].axvline(x=mod_val, color = 'yellow', linestyle='dashed', linewidth = 2, label='Mode')
    ax[0].axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2, label='Maximum')
    ax[0].legend(loc='upper right')

      
    ax[1].boxplot(col, vert=False)
    ax[1].set_xlabel('value', fontsize=10)

    
    fig.suptitle(title, fontsize=15)
```

```python
def show_density(col,baslik):
    from matplotlib import pyplot as plt

    fig = plt.figure(figsize=(15,5))

   
    col.plot.density()

    
    plt.title(baslik, fontsize=15)

    
    plt.axvline(x=col.mean(), color = 'cyan', linestyle='dashed', linewidth = 2, label = 'Mean')
    plt.axvline(x=col.median(), color = 'red', linestyle='dashed', linewidth = 2, label = 'Median')
    plt.axvline(x=col.mode()[0], color = 'yellow', linestyle='dashed', linewidth = 2, label = 'Mode')
    plt.legend()

    
    plt.show()
```

```python
det_skills = df.groupby("restaurant")[['cal_fat', 'total_fat', 'sat_fat','trans_fat']].mean()
det_skills
```

```python
def fatrates(isim):
    firma = det_skills.loc[isim]
    fig, ax = plt.subplots(figsize=(8, 8)) 
    colors = ['#ffbb78','#98df8a','#ff9896','#9467bd','#c5b0d5']
    ax.pie(firma.values, labels=firma.index, autopct='%1.1f%%', colors=colors)
    ax.set_title(f'{isim} Fat Rates')
    plt.show()
```

Bu fonksiyon, det_skills veri çerçevesinde şirketlerin yağ oranlarını görselleştirmek için bir pasta grafiği oluşturur. Fonksiyon, seçilen şirketin yağ oranlarını dilimlerin değerleri olarak kullanır ve dilimlerin etiketlerini belirler. Fonksiyon, oluşturulan grafiği ekranda görüntüler.
